```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/07_scheme_performance.csv")

df.head()
```

```python
top_funds = df.sort_values(
    by="return_5yr_pct",
    ascending=False
).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(
    top_funds["scheme_name"],
    top_funds["return_5yr_pct"]
)

plt.xticks(rotation=90)
plt.title("Top 10 Funds by 5-Year Return")
plt.xlabel("Fund Name")
plt.ylabel("5-Year Return (%)")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["return_5yr_pct"], bins=10)

plt.title("Distribution of 5-Year Returns")
plt.xlabel("5-Year Return (%)")
plt.ylabel("Number of Funds")

plt.show()
```

```python
category_return = df.groupby("category")["return_5yr_pct"].mean()

plt.figure(figsize=(8,5))
category_return.plot(kind="bar")

plt.title("Average 5-Year Return by Category")
plt.xlabel("Category")
plt.ylabel("Average Return (%)")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
top_rated = df.sort_values("morningstar_rating", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    data=top_rated,
    x="morningstar_rating",
    y="scheme_name"
)

plt.title("Top 10 Funds by Morningstar Rating")
plt.xlabel("Morningstar Rating")
plt.ylabel("Fund Name")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    data=df,
    x="category",
    y="expense_ratio_pct"
)

plt.xticks(rotation=45)
plt.title("Expense Ratio Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Expense Ratio (%)")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["aum_crore"],
    bins=20,
    kde=True
)

plt.title("Distribution of AUM")
plt.xlabel("AUM (Crore)")
plt.ylabel("Count")
plt.show()
```
